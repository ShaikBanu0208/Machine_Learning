```python
import seaborn as sns
```

Note: Seaborn has some built-in datasets

```python
# total bill vs tip dataset
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# set a theme
sns.set_theme()
```

```python
# visualise data
sns.relplot(x='total_bill',y='tip',data=tips,col='time',hue='smoker',style='smoker',size='size')
```

Bar Chart

```python
sns.barplot(data=iris,x='petal_length',y='petal_width',hue='species')
```

```python
sns.displot(iris['species'])
```

```python
# Select only the numeric columns for correlation calculation
numeric_iris = iris.select_dtypes(include=np.number)

# Calculate the correlation matrix
correlation = numeric_iris.corr()
```

```python
# constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
